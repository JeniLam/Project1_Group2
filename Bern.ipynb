```python
# Calculation to know the data:

file = "Resources/All_rest_2018.csv"
pop_data = pd.read_csv(file)
pop_data.head(9)
```

# Define Household income - Household income considers the incomes of all people ages 15 years or older occupying the same housing unit, regardless of relation. A single person occupying a dwelling by himself is also considered a household.

# Define Per Capita Income - Per capita income measures the income earned by each person in a given area.  Therefore, two income earners in the same family or household are counted separately when measuring per capita income.


```python
# Count the average Household Income  
average_household_income = pop_data["Household Income"].mean
average_household_income
```

```python
# Count how many stores within each state
state_counts = pop_data["State"].value_counts()
state_counts
```

```python
# Correlation of Restaurants, with regards to population and average income.   

average_income = pop_data["Per Capita Income"].mean()
ttl_population = pop_data["Population"].sum()

ttl_location = pop_data["City"].count()

summary_table = pd.DataFrame({"Per Capita Income": [average_income],
                                                   "Population": [ttl_population],
                             "City": [ttl_location]})
summary_table
```
