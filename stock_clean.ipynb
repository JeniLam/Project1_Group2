```python
#Dependencies
import os
import pandas as pd
```

```python
#Pull in CSV file

file = "Resources/all_stocks_5yr.csv"
stock_df = pd.read_csv(file)
stock_df.head()
```

```python
#Select only ticker symbols for Chipotle, McDonalds, and YUM food groups

stock_clean_df = stock_df.loc[(stock_df['Name'] == "CMG") |
                             (stock_df['Name'] == "MCD") |
                             (stock_df['Name'] == "YUM")]
stock_clean_df.head()
```

```python
#Create new csv with clean data

stock_clean_df.to_csv(r'C:\Users\Dan\gitrepo\Project1_Group2\Resources\stock_clean.csv')
```
